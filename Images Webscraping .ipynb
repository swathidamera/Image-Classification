```python
#import the required libraries
import pandas as pd 
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import ElementClickInterceptedException,NoSuchElementException  

import requests

import warnings
warnings.filterwarnings('ignore')
```

```python
#Connect to the web driver
driver=webdriver.Chrome(r"C:/Users/dravi/Downloads/chromedriver_win32/chromedriver.exe")
```

```python
#get the URL
driver.get('https://www.amazon.in/')
time.sleep(1)
```

```python
#maximise the window
driver.maximize_window()
```

Create empty lists for storing the urls of images of sarees, trosers and jeans

```python
url1=[]
url2=[]
url3=[]
url=[url1,url2,url3]
```

```python

products=['Sarees For women','trousers for men','jeans for men']
for n,prod in enumerate(products):
    
    #entering product in the search field
    search_prod=driver.find_element_by_id('twotabsearchtextbox')
    search_prod.clear()
    search_prod.send_keys(prod)
    
    #clicking the search button
    search_button=driver.find_element_by_id('nav-search-submit-button')
    search_button.click()
    
    im=0
    for i in range(10):
        # Clicking next button to go to the next page
        driver.find_element_by_xpath("//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']").click()
        time.sleep(2)
        
        #Collect the image elements
        images=driver.find_elements_by_xpath("//img[@class='s-image']")

        
        #Collect the image urls and append to url
        for image in images:
            try:
                url[n].append(image.get_attribute('src'))
                im=im+1
                
            except NoSuchElementException:
                pass
         
        if (im >= 200):
            break
           
    
```
